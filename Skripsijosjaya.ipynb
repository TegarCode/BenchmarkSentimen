Case Folding

```python
def case_folding(text):
    if isinstance(text, str):
        lowercase_text = text.lower()
        return lowercase_text
    else:
        return text

df['case_folding'] = df['full_text'].apply(case_folding)
df.head()
```

Cleaning

```python
import re
import string
import nltk

def remove_URL(tweet):
    url = re.compile(r'http://\S+|www\.\S+')
    return url.sub(r'', tweet)

def remove_html(tweet):
    html = re.compile(r'<.*?>')
    return html.sub(r'', tweet)

def remove_numbers_with_words(tweet):
    tweet = re.sub(r'\b\d+[a-zA-Z]+\b', '', tweet)
    tweet = re.sub(r'\b\d+\b', '', tweet)
    return tweet

def remove_link(tweet):
    tweet = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', tweet)
    tweet = re.sub(r' (https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w\.-]*)*\/?\s', '', tweet)
    return tweet

def remove_emoji(tweet):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', tweet)

def remove_symbols(tweet):
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)
    return tweet

def remove_acc(tweet):
    tweet = re.sub(r'@[^\s]+', '', tweet)
    return tweet

# Proses cleaning
df['cleaning'] = df['case_folding'].apply(lambda x: remove_URL(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_html(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_emoji(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_symbols(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_numbers_with_words(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_acc(x))
df['cleaning'] = df['cleaning'].apply(lambda x: remove_link(x))

# Hapus data duplikat berdasarkan kolom 'cleaning'
df = df.drop_duplicates(subset='cleaning').reset_index(drop=True)

# Cek hasil
df.head()
```

```python
import pandas as pd

# Load dataset utama & kamus alay
kamus_alay = pd.read_csv('/content/drive/MyDrive/Skripsi/kamus_alay.csv')  # file yang kamu maksud itu

# Convert ke dictionary
alay_dict = dict(zip(kamus_alay['slang'], kamus_alay['formal']))

# Function normalisasi
def normalisasi_alay(text):
    if isinstance(text, str):
        return ' '.join([alay_dict.get(kata, kata) for kata in text.split()])
    else:
        return text

# Apply ke kolom cleaning
df['normalized'] = df['cleaning'].apply(normalisasi_alay)

df.head()
```

```python
from collections import Counter

# Ambil semua kata sifat (JJ) dari kolom pos_tag
adjectives = []
for tags in df['pos_tag']:
    for word, tag in tags:
        if tag == 'JJ':
            adjectives.append(word)

# Hitung frekuensi kemunculannya
adj_counts = Counter(adjectives)

# Ubah jadi dataframe biar rapi tampilannya
adj_df = pd.DataFrame(adj_counts.items(), columns=['adjective', 'count'])

# Urutkan dari yang paling jarang sampai yang paling sering
adj_df_sorted = adj_df.sort_values(by='count', ascending=True).reset_index(drop=True)

# Tampilkan hasilnya
print(adj_df_sorted)
adj_df_sorted.to_csv('palingsering.csv', index=False)
```

```python
import pandas as pd
import json
import random

# Load dataset hasil preprocessing
df = pd.read_csv('/content/drive/MyDrive/Skripsi/hasil_preprocessingpostag.csv')

# Load tesaurus JSON
with open('/content/drive/MyDrive/Skripsi/dict.json', 'r') as f:
    tesaurus = json.load(f)

# Convert jadi dict mapping: kata → daftar sinonim
kamus_sinonim = {k: v['sinonim'] for k, v in tesaurus.items()}

# Fungsi Synonym Replacement
def synonym_replacement(tokens, alpha=0.3):
    new_tokens = tokens.copy()
    n = max(1, int(alpha * len(tokens)))

    # Pilih indeks kata yang ada di kamus
    candidates = [i for i, token in enumerate(tokens) if token in kamus_sinonim]
    random.shuffle(candidates)

    num_replace = min(n, len(candidates))
    for i in candidates[:num_replace]:
        new_tokens[i] = random.choice(kamus_sinonim[new_tokens[i]])

    return new_tokens

# Dataframe hasil augmentasi
augmented_data = []

# Loop data
for idx, row in df.iterrows():
    tokens = eval(row['Filtering/stopword_removal'])  # pastikan list
    new_tokens = synonym_replacement(tokens, alpha=0.1)
    new_text = ' '.join(new_tokens)

    # Simpan data hasil augmentasi
    new_row = {
        'full_text': new_text,
        'network': row['network'],
        'body': row['body'],
        'display': row['display'],
        'platform': row['platform'],
        'performance': row['performance'],
        'memory': row['memory'],
        'camera': row['camera'],
        'sound': row['sound'],
        'communications': row['communications'],
        'features': row['features'],
        'battery': row['battery'],
        'accessories': row['accessories'],
    }
    augmented_data.append(new_row)

# Convert ke dataframe
df_augmented = pd.DataFrame(augmented_data)

# Simpan hasilnya ke CSV
df_augmented.to_csv('synonym_augmented_dari_tesaurus.csv', index=False)
```

```python
import random
import json
import ast

# Load kamus sinonim JSON
with open('/content/drive/MyDrive/Skripsi/dict.json') as f:
    kamus_sinonim = json.load(f)

# Function untuk replace kata dengan sinonim berdasarkan POS
def synonym_replacement_pos(row, alpha=0.1):
    tokens = ast.literal_eval(row['Filtering/stopword_removal'])
    pos_tags = ast.literal_eval(row['pos_tag'])

    # Ambil kata yang POS-nya di ['JJ', 'VB', 'NN'] dan ada di kamus
    candidate_words = [word for word, pos in pos_tags if pos in ['JJ', 'VB', 'NN'] and word in kamus_sinonim]

    n_replace = max(1, int(alpha * len(tokens)))
    if not candidate_words:
        return ' '.join(tokens)  # kalau gak ada candidate, balikin aja

    words_to_replace = random.sample(candidate_words, min(n_replace, len(candidate_words)))

    new_tokens = []
    for token in tokens:
        if token in words_to_replace:
            sinonim_list = kamus_sinonim[token]['sinonim']
            if sinonim_list:
                new_word = random.choice(sinonim_list)
                new_tokens.append(new_word)
            else:
                new_tokens.append(token)
        else:
            new_tokens.append(token)

    return ' '.join(new_tokens)

# Apply ke dataframe
df['synonym_replacement'] = df.apply(lambda row: synonym_replacement_pos(row, alpha=0.1), axis=1)

# Cek hasil
df[['Filtering/stopword_removal', 'synonym_replacement']].head(10)
```

Antonim

```python
import random
import ast

# Load antonym txt ke dictionary
antonym_dict = {}
with open('/content/drive/MyDrive/Skripsi/antonym_list.txt', 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            word, antonym = parts
            antonym_dict[word] = antonym

# Function untuk antonym replacement berdasarkan POS
def antonym_replacement_pos(row, alpha=0.8):
    tokens = ast.literal_eval(row['Filtering/stopword_removal'])
    pos_tags = ast.literal_eval(row['pos_tag'])

    # Ambil kata yang POS-nya di ['JJ', 'VB', 'NN'] dan ada di kamus antonim
    candidate_words = [word for word, pos in pos_tags if pos in ['JJ', 'VB', 'NN'] and word in antonym_dict]

    n_replace = max(1, int(alpha * len(tokens)))
    if not candidate_words:
        return ' '.join(tokens)  # kalau ga ada candidate, balikin original

    words_to_replace = random.sample(candidate_words, min(n_replace, len(candidate_words)))

    new_tokens = []
    for token in tokens:
        if token in words_to_replace:
            new_word = antonym_dict[token]
            new_tokens.append(new_word)
        else:
            new_tokens.append(token)

    return ' '.join(new_tokens)

# Apply ke dataframe
df['antonym_replacement'] = df.apply(lambda row: antonym_replacement_pos(row, alpha=0.1), axis=1)

# Cek hasil
df[['Filtering/stopword_removal', 'synonym_replacement', 'antonym_replacement']].head(10)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np
import os

# Load dataset
# df = pd.read_csv("data.csv")

# Inisialisasi TF-IDF
vectorizer = TfidfVectorizer(use_idf=True, strip_accents='ascii')
X = vectorizer.fit_transform(df['stemming_data'])

# List aspek sentimen
aspects = ["network", "body", "display", "platform", "performance", "memory",
           "camera", "sound", "communications", "features", "battery", "accessories"]

# Buat folder evaluasi kalau belum ada
os.makedirs("/content/drive/MyDrive/Skripsi/evaluasi/confmatrix_skenario2", exist_ok=True)

# Inisialisasi hasil skenario
results = {}
skenario_2_akurasi = []
akurasi_list = []

for aspect in aspects:
    y = df[aspect]

    # Pembagian Data (70-30)
    X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(X, y, test_size=0.3, random_state=33)

    # Model SVM dengan GridSearchCV
    param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
    grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train_70, y_train_70)

    # Evaluasi Model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test_30)
    accuracy = accuracy_score(y_test_30, y_pred)
    cm = confusion_matrix(y_test_30, y_pred)

    # Simpan confusion matrix ke file CSV
    np.savetxt(f"/content/drive/MyDrive/Skripsi/evaluasi/confmatrix_skenario2/confmatrix_{aspect}.csv", cm, delimiter=",")

    # Simpan hasil ke dict results
    results[aspect] = {
        'best_params': grid_search.best_params_,
        'accuracy': accuracy,
        'classification_report': classification_report(y_test_30, y_pred, output_dict=True)
    }

    # Simpan akurasi ke list skenario
    skenario_2_akurasi.append(accuracy)
    akurasi_list.append({'Aspect': aspect, 'Accuracy': accuracy})

    # Print log ke konsol
    print(f"Aspect: {aspect}")
    print(f"Best Params: {grid_search.best_params_}")
    print(f"Accuracy: {accuracy}")
    print(classification_report(y_test_30, y_pred))
    print("-" * 50)

# Simpan akurasi semua aspek ke CSV
akurasi_df = pd.DataFrame(akurasi_list)
akurasi_df.to_csv("/content/drive/MyDrive/Skripsi/evaluasi/akurasi_skenario2.csv", index=False)

print("Evaluasi skenario 2 selesai dan hasil tersimpan!")
```

```python
def train_svm(X, y, test_size=0.3, use_grid=False, return_model=False):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=33)

    if use_grid:
        param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
        grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
        grid_search.fit(X_train, y_train)
        model = grid_search.best_estimator_
        params = grid_search.best_params_
    else:
        model = SVC(C=1, kernel='linear')
        model.fit(X_train, y_train)
        params = {'C': 1, 'kernel': 'linear'}

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    cm = confusion_matrix(y_test, y_pred)

    if return_model:
        return params, acc, report, model, y_test, y_pred, cm
    else:
        return params, acc, report, y_test, y_pred, cm

# List akurasi
skenario_1_akurasi = []
akurasi_list = []
```
